# _---------- Roll a Dice ----------_ 

```python
#Function to Roll the dice
def rolling(roll_dice):
    #using dictionary to call the funciton based on the Digit on the rolled dice 
    dice_number = {
        1: one,
        2: two,
        3: three,
        4: four,
        5: five,
        6: six
    }
    dice_number.get(roll_dice)()

    
def one():
    return print(r'''
         ___________
        |           |
        |           |
        |     ●     |
        |           |
        |___________|
        ''')


def two():
    return print(r'''
         ___________
        |           |
        |           |
        |   ●   ●   |
        |           |
        |___________|
        ''')

def three():
    return print(r'''
         ___________
        |           |
        |           |
        |  ●  ●  ●  |
        |           |
        |___________|
        ''')

def four():
    return print(r'''
         ___________
        |           |
        |  ●     ●  |
        |           |
        |  ●     ●  |
        |___________|
        ''')


def five():
    return print(r'''
         ___________
        |           |
        |  ●     ●  |
        |     ●     |
        |  ●     ●  |
        |___________|
        ''')

def six():
    return print(r'''
         ___________
        |           |
        |  ●     ●  |
        |  ●     ●  |
        |  ●     ●  |
        |___________|
        ''')

# Generating random number
while True:
    roll_quit = input("Enter 'R' to roll the dice or 'Q' to quit the dice = ").upper()
    if roll_quit == 'R':
        import random
        print("Rolling Dice...")
        roll_dice = random.randint(1,7)
        rolling(roll_dice)
    elif roll_quit == 'Q':
        print("Quit Dice")
        import sys
        sys.exit()
    else:
        print("Wrong input!")
```
